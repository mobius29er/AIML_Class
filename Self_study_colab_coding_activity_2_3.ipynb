## Data analysis

### Subtask:
Perform a more in-depth analysis of the Gapminder dataset, focusing on relationships between variables and potential outliers.


**Reasoning**:
Perform correlation analysis, outlier detection, grouped analysis, and temporal analysis as instructed.  Visualize the results using heatmaps and boxplots.



````python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Correlation Analysis
numerical_cols = ['lifeExp', 'pop', 'gdpPercap']
correlation_matrix = gap_df[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()

# 2. Outlier Detection
plt.figure(figsize=(12, 6))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x='continent', y=col, data=gap_df)
    plt.title(f'Box Plot of {col} by Continent')
plt.tight_layout()
plt.show()

# 3. Grouped Analysis
grouped_stats = gap_df.groupby('continent')[numerical_cols].agg(['mean', 'median', 'std'])
display(grouped_stats)


# 4. Temporal Analysis
temporal_analysis = gap_df.groupby(['continent', 'year'])[numerical_cols].mean().reset_index()

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i+1)
    for continent in gap_df['continent'].unique():
      subset = temporal_analysis[temporal_analysis['continent'] == continent]
      plt.plot(subset['year'], subset[col], label=continent)
    plt.xlabel('Year')
    plt.ylabel(col)
    plt.title(f'Trend of {col} over time')
    plt.legend()
plt.tight_layout()
plt.show()
````

## Data visualization

### Subtask:
Visualize the key characteristics of the Gapminder dataset.


**Reasoning**:
Create histograms, box plots, and scatter plots to visualize the key characteristics of the Gapminder dataset, as instructed. Use logarithmic scales for 'pop' and 'gdpPercap' where appropriate.



````python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.hist(gap_df['lifeExp'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.title('Distribution of Life Expectancy')

plt.subplot(1, 3, 2)
plt.hist(gap_df['pop'], bins=20, color='salmon', edgecolor='black', log=True)
plt.xlabel('Population')
plt.ylabel('Frequency (log scale)')
plt.title('Distribution of Population')

plt.subplot(1, 3, 3)
plt.hist(gap_df['gdpPercap'], bins=20, color='lightgreen', edgecolor='black', log=True)
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency (log scale)')
plt.title('Distribution of GDP per Capita')

plt.tight_layout()
plt.show()

# Box plots
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.boxplot(x='continent', y='lifeExp', data=gap_df, palette='Set3')
plt.xlabel('Continent')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy by Continent')

plt.subplot(1, 3, 2)
sns.boxplot(x='continent', y='pop', data=gap_df, palette='Set3', showfliers=False)  # Remove outliers for better visualization
plt.yscale('log')
plt.xlabel('Continent')
plt.ylabel('Population (log scale)')
plt.title('Population by Continent')

plt.subplot(1, 3, 3)
sns.boxplot(x='continent', y='gdpPercap', data=gap_df, palette='Set3', showfliers=False)  # Remove outliers for better visualization
plt.yscale('log')
plt.xlabel('Continent')
plt.ylabel('GDP per Capita (log scale)')
plt.title('GDP per Capita by Continent')

plt.tight_layout()
plt.show()


# Scatter plots
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.scatterplot(x='gdpPercap', y='lifeExp', hue='continent', data=gap_df, palette='viridis', s=50)
plt.xscale('log')
plt.xlabel('GDP per Capita (log scale)')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs. GDP per Capita')

plt.subplot(1, 3, 2)
sns.scatterplot(x='pop', y='lifeExp', hue='continent', data=gap_df, palette='viridis', s=50)
plt.xscale('log')
plt.xlabel('Population (log scale)')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs. Population')

plt.subplot(1, 3, 3)
sns.scatterplot(x='pop', y='gdpPercap', hue='continent', data=gap_df, palette='viridis', s=50)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Population (log scale)')
plt.ylabel('GDP per Capita (log scale)')
plt.title('GDP per Capita vs. Population')

plt.tight_layout()
plt.show()
````
